```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as npR
import seaborn as sns
```

```python
custom_palette = ["#912BBC", "#D875C7", "#E9A89B", "#FFEBB2"]
sns.set_palette(custom_palette)
```

```python
pd.set_option("display.max_columns",24)
```

```python
dataset = pd.read_excel("US Superstore data.xlsx")
```

```python
# dataset.head(2)
```

```python
# dataset.isnull().sum()
```

# Categories item

```python
dataset["Category"].value_counts()
```

```python
plt.figure(figsize = (6,4))
sns.countplot(x = "Category", data = dataset)
plt.title("count of categories")
plt.savefig("count of categories.jpg")
plt.show()
```

# Ship Mode

```python
dataset["Ship Mode"].value_counts()
```

```python
x = dataset["Ship Mode"].value_counts().index
y = dataset["Ship Mode"].value_counts().values
```

```python
plt.figure(figsize=(4,4))
plt.pie(y,labels=x, startangle = 60, autopct = "%0.2f%%")
plt.title("count of categories")
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.savefig("count of Ship Mode.jpg")
plt.show
```

# Customer Segment

```python
plt.figure(figsize=(4,4))
sns.countplot(x = "Segment", data = dataset)
plt.title("Customer Segment")
plt.savefig("CustomerSegment.jpg")
plt.show()
```

# Ship Modes by Segment and Profit Insights

```python
plt.figure(figsize=(6,4))
sns.barplot(x="Category", y="Profit", data=dataset, estimator=sum, ci=None, hue = "Segment", errorbar = None)
plt.title("Ship Modes by Segment and Profit Insights")
plt.savefig("Ship Modes by Segment and Profit Insights.jpg")
plt.show()
```

# Ship Mode Utilization in Product Categories

```python
plt.figure(figsize=(6,4))
sns.countplot(x = "Ship Mode", data = dataset, hue = "Category")
plt.title("Distribution of Ship Modes by Product Category")
plt.savefig("Distribution of Ship Modes by Product Category.jpg")
plt.show()
```

# Sub-Category Distribution within Office Supplies

```python
plt.figure(figsize=(6,4))
sns.countplot(x = "Category", data = dataset[dataset["Category"]=="Office Supplies"], hue = "Sub-Category")
plt.title("Distribution of Office Supplies by Sub-Category")
plt.savefig("Distribution of Office Supplies by Sub-Category.jpg")
plt.show()
```
